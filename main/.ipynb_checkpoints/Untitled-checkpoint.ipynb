```python
trans = MICE(verbose=False)
```

```python
fitted = trans.fit(dta.drop(['DATE', 'PRICE(Cents/kWh.5)'], axis= 1))
```

```python
transformed = fitted.transform(dta.drop(['DATE', 'PRICE(Cents/kWh.5)'], axis= 1))
```

```python
xxx= pd.DataFrame(transformed)
xxx.columns = ['GENERATION\n(Megawatthours)', 'Coal (Short Tons)',
       'Natural Gas (Mcf)', 'Other Gases (Billion BTU)', 'Petroleum (Barrels)',
       'internet_search', 'AverageTemperature', 'Precipitation', 'covid','coal_price',
       'naturalgas_price', 'petroleum_price', 'nuclear_generation']
```

```python
xxx['DATE']= pd.to_datetime(dta['DATE'])
```

```python
xxx
```

```python
(
    ggplot(xxx,aes(x="DATE", y = "price")) +
    geom_line()
)
```

```python
(
    ggplot(xxx,aes(x="DATE", y = 'coal_price')) +
    geom_line()
)
```
